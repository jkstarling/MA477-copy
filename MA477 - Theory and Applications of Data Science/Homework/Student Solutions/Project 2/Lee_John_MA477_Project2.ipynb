# Analysis

   In this project, we are tasked with modeling the probability of airline customer satisfaction based on numerous explanatory variables. The explanatory variables are all related to customer air-travel experience in general, and range from characteristics like age to amenities like baggage handling. It is crucial to note that our response variable, airline satisfaction,is categorical rather than quantitative. This means customers can only be satisfied or neutral/dissatisfied. This indicates that this data is best modeled through classification rather than regression. 
   
   We first wrangle the data into a format we can deal with. We use data visualization techniques we verified that nearly all our variables exhibited linear relationships with satisfaction, with the only exceptions being gate location and gender. This bodes well for potential model-building, and we press on with all our variables since we know that logistic regression and other classification methods can shrink non-useful variables anyways. We transform all qualitative variables into quantitative variables through assigning dummy variables. We then note that our data has several thousand missing values across several columns (out of over 100,000 total rows). We address this by imputing values into the 'type of travel' variable based on the similar 'class' variable, removing all rows where the response variable is null, and then imputing the average value of each column into any other missing values. This allows for us to retain most of our large data set (still clearing 100,000 observations) while filling NA values with reasonable substitutes. Next, we standardize all our explanatory variables due to their varying units, before finally splitting off our response into a separate data frame. We select three classification algorithms for modeling: Logistic Regression, K-Nearest Neighbors (Classifier), and Gaussian Naive Bayes. Cross Validation with default parameters reveals KNN as the model with both the highest accuracy and recall. However, we tune model parameters in Logistic Regression (C, penalty) and KNN (neighbors) to confirm this notion. Our results show the model which optimizes accuracy and recall is KNN(neighbors = 6), with cross-validated values of 0.92 and 0.97 respectively. to explore prediction power, we reinforce our high recall and accuracy values of our selected model with a ROC curve which shows an impressive 0.97 AUC, demonstrating our models strength in differentiating between classes. Variation of our model is explored with standard deviation of our selected method's cross validation accuracy, which we find to be just 0.016 which is relatively small compared to our accuracy values themselves. Thus we conclude our model has low variance. Lastly, we explore inference and come to conclusions on which factors have the highest correlation with satisfaction through examining standardized coefficients of our fitted model. The value with the highest positive association with satisfaction is larger values of online boarding scores, while the greatest negative correlation is when customers are on personal rather than business travel. 

# Data Wrangling

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_validate, cross_val_score, ShuffleSplit,  KFold, train_test_split
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score, recall_score, roc_curve, auc
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
import statsmodels.api as sm
```

After reading in our data, we must first examine our data for missing values and inspect the variables. We check for how many null values first. It quickly becomes apparent that we have thousands of missing values across several of our columns.

```python
airline = pd.read_csv('airline_train.csv')
airline
```

```python
airline.isnull().sum().sum()
```

```python
bool_series = pd.isnull(airline["Type of Travel"])
airline[bool_series]
```

One column with almost 5000 missing values is Type of Travel. We impute into this column based on similarly named values in Class. The only question is how to impute missing Eco Plus rows. We find that about 300 more Eco Plus entries are with Business Travel than Personal Travel, so we imput Business Travel. Thus, for NA values of 'type of travel' that have 'class' Eco we impute Personal travel, while for values with 'class' Eco plus or Business we impute Business Travel. 

```python
airEcoPlus = airline[airline['Class'] == 'Eco Plus']
airEcoPlusPersonal =airEcoPlus[airEcoPlus['Type of Travel'] == 'Personal Travel']
airEcoPlusPersonal
```

```python
airEcoPlusBusiness =airEcoPlus[airEcoPlus['Type of Travel'] == 'Business travel']
airEcoPlusBusiness
```

```python




airline['Type of Travel'] = np.where(airline['Type of Travel'].isnull(),airline['Class'],airline['Type of Travel'])

airline['Type of Travel'] = np.where(airline['Type of Travel'] == "Eco","Personal Travel",airline['Type of Travel'])

airline['Type of Travel'] = np.where(airline['Type of Travel'] == "Business","Business travel",airline['Type of Travel'])

airline['Type of Travel'] = np.where(airline['Type of Travel'] == "Eco Plus","Business travel",airline['Type of Travel'])
```

```python
airline['Type of Travel'].isnull().sum()
```

We want to drop any entries that have missing values for our response variable, since it doesn't make sense to impute variables into our respons

```python
airline1 = airline[airline['satisfaction'].notna()]
```

Our next task is to deal with qualitative variables. All of our qualitative variables have just two levels except for 'Class' which has three. Thus we change our qualitative variables to 0s and 1s with the exception of the 'Class' variable which we assign 0,1, and 2 from lowest (Eco) to highest (Business). 

```python
def func(x):
    if x == "Eco":
        return 0
    elif x == "Eco Plus":
        return 1
    else:
        return 2
```

```python
airline_new=pd.get_dummies(airline1,columns=['Gender','Customer Type','Type of Travel','satisfaction'],drop_first=True)
airline_new['Class'] = airline_new['Class'].apply(func)
```

```python
airline_new
```

Lastly, we deal with our remaining missing values through imputing the mean of each column

```python
airline_fill = airline_new.fillna(airline_new.mean())
airline_fill
```

# Data Normalization and Separation  

Our final data manipulation task is the question of standardizing our variables. Though we have a good amount of categorical variables which need not be standardized, we have sevaral quantitative explanatory variables with very differnent units. Thus we simply standardize all of our explanatory variables to avoid high variability in our models. We then separate off our response variable satisfaction into its own data frame. Note that we also drop the 'ID' variable which is not needed.

```python
X = airline_fill.drop(['satisfaction_satisfied','id'],axis = 1)
y=airline_fill['satisfaction_satisfied']
```

# Model Evaluation

Since we are faced with a classification problem (our response variable is categorical), we select three suitable methods: Logistic Regression, KNN Classifier, and Gaussian Naive Bayes. With default parameters, we cross validate our data into 10 folds before evaluating mean accuracy and recall for each method. We find that KNN has the highest accuracy (0.92) and recall for unsatisfied customers (0.96). 

```python
log = LogisticRegression(solver='liblinear')
knn = KNeighborsClassifier()
nb = GaussianNB()
```

```python
folds = np.arange(1,10,1)

accuracy_list = []
recall_list = []

for i in folds:
    X_train,X_test,y_train,y_test=train_test_split(X_sc,y,test_size=0.3,random_state=i)
    log.fit(X_train,y_train)
    pred=log.predict(X_test)
          
    accuracy_list.append(accuracy_score(y_test,pred))
    recall_list.append(recall_score(y_test,pred,pos_label=0))

```

# Model Tuning

Results can vary substantially depending on which parameters are used in each model, so we investigate our initial findings through tuning the Logistic Regression and KNN models. For logistic regression, we manually check each of the main regularization (penalty) parameters l1 and l2 which specify the norm used in penalization, in addition to looping through 40 values of C. The smaller C is, the stronger the regularization. We find that accuracy and recall roughly stay at 0.83 and 0.91 respectively regardless of parameters.

```python
X_train,X_test,y_train,y_test=train_test_split(X_sc,y,test_size=0.3,random_state=101)
```

```python
c_list = np.arange(0.1,30,0.5)

accuracy_log = []
recall_log = []
auc_log = []


for c in c_list:
    log = LogisticRegression(penalty='l2',solver='liblinear',C=c)
    log.fit(X_train,y_train)
    log_predict = log.predict(X_test)
    log_predict_prob = log.predict_proba(X_test)
    tp,fp,_=roc_curve(y_test,log_predict_prob[:,0])
        
    roc_auc = auc(fp,tp)
    recall_unsatisfied_Log = recall_score(y_test,log_predict,pos_label=0)
    accuracy_unsatisfied_Log = recall_score(y_test,log_predict)
        
    auc_log.append(roc_auc)
    recall_log.append(recall_unsatisfied_Log)
    accuracy_log.append(accuracy_unsatisfied_Log)
        
```

```python
sns.set_style('whitegrid')
fig=plt.figure(figsize=(8,6))

axes=fig.add_axes([0,0,1,1])

axes.plot(c_list,accuracy_log, linestyle='-.',lw=1.5,label='accuracy')
axes.plot(c_list,recall_log, linestyle='-.',lw=1.5,label='recall')
axes.plot(c_list,auc_log, linestyle='-.',lw=1.5,label='recall')

axes.set_xlabel("C",fontsize=14)
axes.set_ylabel("Mean accuracy/recall Score from CV",fontsize=14)

plt.legend(fontsize=12)
plt.show()
```

In the case of KNN, we loop over the value of 'k' which specifies how many neighbors to consider before classifying an observation. We test values from 2 to 20, and find that accuracy and recall are optimized at approximately k = 6.

```python
k_list = np.arange(2,20,2)

accuracy_knn = []
recall_knn = []
auc_knn = []


for k in k_list:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train,y_train)
    knn_predict = knn.predict(X_test)
    knn_predict_prob = knn.predict_proba(X_test)
    tp,fp,_=roc_curve(y_test,knn_predict_prob[:,0])
        
    roc_auc = auc(fp,tp)
    recall_unsatisfied_KNN = recall_score(y_test,knn_predict,pos_label=0)
    accuracy_unsatisfied_KNN = recall_score(y_test,knn_predict)
        
    auc_knn.append(roc_auc)
    recall_knn.append(recall_unsatisfied_KNN)
    accuracy_knn.append(accuracy_unsatisfied_KNN)
```

```python
auc_knn
```

```python
sns.set_style('whitegrid')
fig=plt.figure(figsize=(8,6))

axes=fig.add_axes([0,0,1,1])

axes.plot(k_list,accuracy_knn, linestyle='-.',lw=1.5,label='accuracy')
axes.plot(k_list,recall_knn, linestyle='-.',lw=1.5,label='recall')
axes.plot(k_list,auc_knn, linestyle='-.',lw=1.5,label='auc')

axes.set_xlabel("K Neighbors",fontsize=14)
axes.set_ylabel("Mean accuracy/recall Score from CV",fontsize=14)

plt.legend(fontsize=12)
plt.show()
```

After tuning the Logistic Regression and KNN models, we see that KNN appears to maximize accuracy and recall at k = 6 neighbors.
