## Data Preparation

#### Task 1 and 3:

#### After reading in the data, we underwent a series of steps to clean the data, filling in missing values and converting qualitative explanatory variables. First, we removed three excess columns from the data set. This included two unnamed index columns and the ID column, all three of which played no meaningful role in a customer's satisfaction level. Next, we inspected missing data using both a heatmap and the .info() method. We found that six columns had missing data: Type of travel, Inflight entertainment, baggage handling, departure delay, Arrival delay, and satisfaction. Using our data analysis to inform our decisions, we decided to drop all rows missing data in the Type of Travel and satisfaction columns. This was done because Type of Travel was both a two-level qualitative variable and did not seem to correlate closely with another variable (found via data analysis), making it tough to impute accurate values. Missing rows in satisfaction were dropped because this is the response variable (containing the targets we are trying to predict), meaning that we had to handle it in a fragile manner. Additionally, it was missing only around 3% of the data, a small portion. 

#### Next, we converted all qualitative string values to 0/1 dummy variables for five columns. Finally, for Inflight entertainment, baggage handling, departure delay, and Arrival delay we imputed missing values with the mean of each column. This decision was made because all four of these variables had values between 1-5 as a rating scheme. This made mean a more feasible estimate as compared to binary variables like Type of Travel. These variables were also missing a considerable amount of data (around 10%), making imputation necessary in order to preserve data.

#### Tune Parameters for logistic regression; plot shows accuracy and recall are constant for varying c values

```python
def custom_predict(pred,thresh):
    predictions=[]
    for item in pred:
        if item>thresh:
            predictions.append(0)
        elif item<=thresh:
            predictions.append(1)
    return predictions
```

```python
from collections import defaultdict
from sklearn.metrics import accuracy_score

```

```python
reg={'l1','l2'}
param_c=np.arange(0.1,5,0.1)

def tune_log_reg(param_c,reg,thresh=0.5,iterations=10):

    
    recall=defaultdict(list)
    accuracy=defaultdict(list)
    
    for penalty in reg:
        for c in param_c:
            recall_c=[]
            accuracy_c=[]
            auc_c=[]
            for i in range(iterations):
                X_train,X_test,y_train,y_test=train_test_split(X_sc,y,test_size=0.3,random_state=i+1)
                
                
                lg=LogisticRegression(penalty=penalty,C=c,solver='liblinear')
                lg.fit(X_train,y_train)
                pred=lg.predict_proba(X_test)
                pred=custom_predict(pred[:,0],thresh)
                #print(pred)
                #print(recall_score(y_test,pred,pos_label=0))
                recall_c.append(recall_score(y_test,pred,pos_label=0))
                accuracy_c.append(accuracy_score(y_test,pred))
            #print(recall_c)
            recall[penalty].append(np.array(recall_c).mean())
            accuracy[penalty].append(np.array(accuracy_c).mean())
    return recall,accuracy
```

```python
recall,accuracy=tune_log_reg(param_c,reg,iterations=100,thresh=0.32)
```

```python
plt.figure(figsize=(10,6))
plt.plot(param_c,recall['l1'],'r--',label='Recall l1')
plt.plot(param_c,recall['l2'],'b-.',label='Recall l2')
plt.plot(param_c,accuracy['l1'],'r-',label='Accuracy l1')
plt.plot(param_c,accuracy['l2'],'b-',label='Accuracy l2')
plt.yticks(np.arange(.82,.985,0.02))
plt.xticks(np.arange(0,5,0.15))
plt.xlim(0,3)
plt.ylim(0.82,0.98)
plt.xlabel('C',fontsize=14)
plt.legend(fontsize=12)
plt.show()
```
