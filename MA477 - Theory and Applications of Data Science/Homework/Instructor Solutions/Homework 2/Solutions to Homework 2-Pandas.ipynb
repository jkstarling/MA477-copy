<h4> Find all the listings of room type <font color='red'> Shared rooms</font> in <font color='red'> Manhattan</font> with a price range of <font color='red'> \$60$-$\$75</font> and that have received over <font color='red'>100 reviews</font></h4>

```python
#Enter your answer here
abnb[(abnb['room_type']=='Shared room') & (abnb['neighbourhood_group']=='Manhattan')
     &(abnb['price'].isin(list(range(60,75)))) &(abnb['number_of_reviews']>100)]
```

<h4> Find the <font color='red'>minimum</font>, <font color='red'>maximum</font>, and <font color='red'>average</font> values for both <font color='red'>price</font> and <font color='red'> number of reviews</font> for the listings in each of the different neighbourhoods of Manhattan</h4>

```python
#Your answer
abnb[abnb['neighbourhood_group']=='Manhattan'].groupby('neighbourhood').agg({'price':[np.min,np.mean,np.max],'number_of_reviews':[np.min,np.mean,np.max]})
```

<br>

Find all the listings that have any version (upper case, lower case, mix etc.) of the following two phrases <b> Females Only</b> and <b> Comfy Couch</b> in the <font color='red'>name</font> column of the <b> abnb</b> dataframe. Display only the <b>id, name, host_name, price,</b> and <b> neighbourhood_group</b>

```python
#Don't change this
pd.set_option('display.max_colwidth', -1)
```

```python
import re
```

```python
pattern='.*females only.*|.*comfy couch.*'
```

```python
females_couch=[]
for index,item in zip(abnb['name'].index,abnb['name'].values):
    if re.search(pattern,str(item).lower()):
        #print('Name:{}\n Index:{}'.format(item,index))
        females_couch.append(index)
 
```

```python
abnb.loc[females_couch][['id','name','host_name','price','neighbourhood_group']]
```
