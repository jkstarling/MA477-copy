<h3> N-Grams</h3>

Often times when analyzing text it is not enough to look at single words, because as we know, in many cases the meaning of the word may change drastically depending on the context. As a simple example take the words <b> good, bad, sorry, great </b> etc. Taken out of context and without looking at neighboring words then <b> good </b> has a positive connotation while <b> bad</b> has a negative connotation. But what if these words were taken ou of the following sentences:

            MA477 is not bad at all!
            
            My social science class was not good at all! 
            
If we were trying to determine the overall sentiment of these two sentences, unless we look at the neighboring words of the key adjectives <b> good, bad</b> we would most likely misjudge the sentiment. For example, we may classify the first statement as having a highly negative sentiment, while the second statement as having a neutral sentiment, where it's clear that the first statement has a positive and the second a highly negative sentiment. 

This is one of the main reasons why we often need to look at two, three, or even more words at the same time to determine the sentiment of a review, statement, etc. This is exactly where the idea of $n-$grams stems from. For example, a $2-$gram is where we break a sentence into pairs of words, a $3-$gram refers to breaking up a sentence/text into tuples of three, so that we can factor in the context in which a specific word is being used. 



```python
import nltk
```

```python
nltk.download('gutenberg')
```

```python
text_docs=nltk.corpus.gutenberg.fileids()
```

```python
text_docs
```

```python
alice=nltk.corpus.gutenberg.sents('carroll-alice.txt')
```

```python
text=alice[5]
```

```python
text
```

So, say instead of wanting to look at each word separately, we want to look at pairs of words. We can do so by applying the `bigrams()` method:


```python
bigrams=nltk.bigrams(text)
```

```python
for item in bigrams:
    print(item)
```

Similarly, for better context, we may want to look at 3-tuples of words:

```python
trigram=nltk.trigrams(text)
```

```python
for item in trigram:
    print(item)
```

As you may imagine, there is a more general method that may be used to form any n-tuple of words.

```python
from nltk.util import ngrams
```

```python
tuple_4=ngrams(text,4)
```

```python
for item in tuple_4:
    print(item)
```

<h3>Regular Expressions (RegEx)</h3>

For a much more detailed discussion of RegEx see https://docs.python.org/3.4/library/re.html.

A <b>regular expression</b>, often abbreviated as <b> RegEx</b>, is a special sequence of characters used to search for, check, and find other strings that fit a specific pattern in a text. For example, we can use RegEx to find all words in a document that begin with 'us' or end in 'ul', or extract only email addresses or phone numbers from a document.

For example, it may be of interest to extract only the names of managers, or coaches, or only the email addresses, or only the phone numbers, or a combination of the above. RegEx comes to rescue!

```python
import re
```

```python
with open('Coaches_text0.txt', 'r') as file:
    coaches_file=file.read()
```

```python
print(coaches_file)
```

<h3>Task</h3>

Suppose we want to find all the email addresses. How would we do that?

Let's start by thinking of the pattern that an email address fits.

```python
pattern='[\w\.\d-]+@[\w\.-]+'

emails=re.findall(pattern,coaches_file)
```

```python
emails
```
