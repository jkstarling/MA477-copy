 In this lecture we will demonstrate that using K-Means clustering (or just a clustering method in general) as a data pre-processing tool can significantly improve the predictive power of the model. In other words, clustering the data into different subgroups before fitting predictive model, often, may result in a significant reduction of the error rate (for classification problems) or MSE score (for regression problems).
 
 In this lecture we will use a pipeline to streamline the data transformation with the end step being a Logistic Regression Classifier.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_validate, KFold
import random
```

```python
airline=pd.read_csv('airline_train.csv',index_col=[0])
airline.drop(airline.columns[:2],inplace=True,axis=1)
airline.dropna(inplace=True)
airline.reset_index(inplace=True,drop=True)
```

```python

rows=random.sample(list(airline.index),15000)
airline=airline.loc[rows]

airline.reset_index(inplace=True,drop=True)
airline['satisfaction']=airline['satisfaction'].apply(lambda x: 1 if x=='satisfied' else 0)
X=airline.drop('satisfaction',axis=1)
y=airline['satisfaction']
```

```python
#Convert qualitative variables to numerical dummy variables

cols=['Gender','Customer Type', 'Type of Travel','Class']

X=pd.get_dummies(data=X,columns=cols,drop_first=True)
```

```python
X.head()
```

<h2> Validation Set</h2>

```python
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.2,random_state=11)
```

```python
scaler=StandardScaler()

scaled_train=scaler.fit_transform(X_train)
X_train_sc=pd.DataFrame(scaled_train,columns=X_train.columns,index=X_train.index)


scaled_val=scaler.fit_transform(X_val)
X_val_sc=pd.DataFrame(scaled_val,columns=X_val.columns,index=X_val.index)
```

```python
X_train_sc.head()
```

```python
X_val_sc.head()
```

<h2>Baseline</h2>

We will fit a Logistic Regression and Random Forest, separately, to create a baseline for model performance.

```python
from sklearn.linear_model import LogisticRegression
from collections import defaultdict
from sklearn.metrics import accuracy_score
```

```python
penalties=['l1','l2']
param_c=np.arange(0.01,2,0.1)

kf=KFold(n_splits=10,shuffle=True,random_state=11)

results=defaultdict(dict)

for reg in penalties:
    reg_dict={}
    for c in param_c:
        
        lg_cv=cross_validate(LogisticRegression(penalty=reg,C=c,solver='liblinear'),X_train_sc,y_train,cv=kf)
        reg_dict['param_'+str(c)]=lg_cv['test_score'].mean()
    results[reg]=reg_dict
        
        
```

```python
df=pd.DataFrame(results)
```

```python
sns.set_style('whitegrid')
df.plot(figsize=(8,6))
```

```python
lg=LogisticRegression(solver='liblinear',C=0.51,penalty='l2')
lg.fit(X_train_sc,y_train)
```

```python
pred_lg=lg.predict(X_val_sc)
```

```python
accuracy_score(y_val,pred_lg)
```

<h2>Pipeline with K-Mean Clustering and LogReg</h2>

```python
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
```

 In a pipeline we sequentially apply a list of transforms with a final estimator. In other words, all of the intermediate steps in a pipeline must be some kind of transformation of the data; that is, they cannot be estimators, except for the last step.
 
 One of the main advantages of using a pipeline is so that we can tune the model via cross-validation and grid search all at once.

```python
clusters=np.arange(12,19,2)
penalties=['l1','l2']
param_c=np.arange(0.1,2,0.2)

kf=KFold(n_splits=5,shuffle=True,random_state=11)

results_pip=defaultdict(dict)

for k in clusters:
    clust={}
    for c in param_c:
        
        pipeline=Pipeline([('scaler',StandardScaler()),
                           ('kmeans',KMeans(n_clusters=k)),
                           ('lg',LogisticRegression(solver='liblinear',C=c,penalty='l2'))])
        
        cv_pip=cross_validate(pipeline,X_train,y_train,cv=kf)
        clust[c]=cv_pip['test_score']
#         print(clust[c])
    results_pip['clust_'+str(k)]=clust
#     print(results_pip['clust_'+str(k)])
    
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(20,30))
m=len(results_pip.keys())
n=len(results_pip['clust_'+str(12)].keys())
for i,k in enumerate(clusters,start=1):
    #min_val=results_pip['clust_'+str(k)][].min()
    plt.subplot(m,2,i)
    sns.boxplot(data=pd.DataFrame(results_pip['clust_'+str(k)]))
    plt.xticks(range(0,n),[x.round(3) for x in results_pip['clust_'+str(k)].keys()])
    plt.yticks(np.arange(0.85,0.890,0.005),fontsize=14)
    plt.xlabel('C',fontsize=14)
    plt.ylabel("Accuracy",fontsize=14)
    plt.title("# clusters {}".format(k),fontsize=18)
    
plt.tight_layout()
```

```python
pipeline=Pipeline([
    ('scaler',StandardScaler()),
    ('kmeans',KMeans(n_clusters=18)),
    ('lg',LogisticRegression(solver='liblinear',C=0.3,penalty='l2'))
])
```

```python
pipeline.fit(X_train,y_train)
```

```python
pipeline.score(X_val,y_val)
```

<h2>Grid Search</h2>

Python has a few built-in methods to search for best parameters. In one of our previous lectures we mentioned RandomSearchCV, another great method is GridSearchCV, which we implement below.

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid=dict(kmeans__n_clusters=np.arange(26,38,3),lg__C=np.arange(0.1,2,0.2))
```

```python
pipeline=Pipeline([
    ('scaler',StandardScaler()),
    ('kmeans',KMeans()),
    ('lg',LogisticRegression(solver='liblinear',penalty='l2'))
])
clf_grid=GridSearchCV(pipeline,param_grid,cv=5,verbose=2)
clf_grid.fit(X_train,y_train)
```

```python
clf_grid.best_score_
```

```python
clf_grid.best_params_
```

<h4>Model performance on the validation set</h4>

```python
model=clf_grid.best_estimator_
```

```python
model.score(X_val,y_val)
```

By using K-Means Clustering as a pre-processing tool we have managed to reduce the error rate by around $21\%$ over the base model; that is, the error rate went from $12.77\%$ to around $10.017\%$. This is a significant improvement over the Logistic Regression alone.
