<h2>Image Segmentation</h2>

Image segmentation is the process of partitioning an image into separtae segments. Semantic segmentation is widely used in self-driving cars. All pixcels that belong to the same object type are assigned to the same image segment. For example, all pixels belonging to the street may be assigned to the same <b>street</b> image segment, or all pixels belonging to pedestrians may be assigned to the <b> pedestrian</b> image segment. 

Semantic segmentation is used to partition the image into regions of interest; that is, it is a way to identify the objects that we would like to analyze further and the background.


Semantic segmentation (especially instance segmentation) nowadays is performed with the state of the art deep neural nets. In this lecture we will simply illustrate the idea of image segmentation via K-Means clustering by performing <b>color segmentation</b>. In other words, we will assign all the pixels that have a similar color to the same image segment.
